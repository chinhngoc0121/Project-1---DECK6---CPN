```python
import pandas as pd
import matplotlib.pyplot as plt
import_data = pd.read_csv('Ecommerce_Purchases.csv')
ecom_data = pd.DataFrame(import_data)
ecom_data   
```

```python
jobs_raw= ecom_data['Job'].drop_duplicates()
print(jobs_raw)
```

```python
result = ecom_data['Job']
result.to_csv("output.csv")
print(type(result))
```

```python
#clean Jobs
temp =[]
for job in result:
    job = job.replace(",","").replace(")","").replace("(","")
    job = str.lower(job)
    job = job.split(" ")
    job.sort()
    temp.append(" ".join(job))

temp_df = pd.DataFrame(temp)
temp_df.to_csv("my_list.csv",index=False,header=False)
print(temp)



job_prc = pd.DataFrame({'Job':temp,'Purchase Price': ecom_data['Purchase Price'].tolist()})
```

```python
#question1 Find the relationship between Job designation and mean Purchase amount
total_prc = job_prc.groupby("Job")["Purchase Price"].sum().reset_index()
top5_job = total_prc.nlargest(5,'Purchase Price')

plt.figure(figsize=(10, 6))
plt.plot(top5_job["Job"], top5_job["Purchase Price"])
plt.xlabel("Job")
plt.ylabel("Mean Purchase Price")
plt.title("Relationship Between Top 5 Job Designation and Total Purchase Price")
plt.xticks(rotation=45)  
plt.show()
```
